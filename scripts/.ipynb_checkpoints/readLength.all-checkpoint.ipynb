```r
library(Biostrings)
library(ggplot2)
library(dplyr)
```

```r
BGI.allfq = readDNAStringSet("../BGI_fastq/MT44_01/TB20003420-202401221650300_read.fq.gz",format = "fastq")
```

```r
BGI.allfq.table = table(cut(width(BGI.allfq),breaks = c(seq(0,20000,500),Inf)))
```

```r
range.df=  data.frame(readLength = factor(names(BGI.allfq.table),levels=names(BGI.allfq.table)),count = as.vector(BGI.allfq.table))
```

```r
options(repr.plot.height = 6,repr.plot.width = 10)
range.df %>% ggplot(aes(x = readLength,y = count)) + geom_bar(stat = "identity") + theme_classic()+ theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))+ggtitle("Cyclone")
```
