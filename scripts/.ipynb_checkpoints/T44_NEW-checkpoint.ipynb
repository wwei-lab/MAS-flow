```r
sci_fq_T44 = readDNAStringSet("../fastq/FAX23635_pass_12b693ce_c74c2dc4.fastq",format = "fastq")
fqnames_T44 = unlist(mclapply(str_split(names(sci_fq_T44)," "),`[`,1,mc.cores = 40))
```

```r
fqnames_T44 = read.table("../fastq/FAX23635_pass_12b693ce_c74c2dc4.fastq.names")
fqnames_T44 = fqnames_T44$V1
```

```r
bed_T44 = import("../alignments/FAX23635_pass_12b693ce_c74c2dc4.fastq.sorted.bed")
bed_T44.names = bed_T44 %>% filter(score >= 6)
bed_T44.names = bed_T44.names$name
```

```r
# adapter.files.T44.mismatch0 = list.files("../data/FAX23635_pass_12b693ce_c74c2dc4/",pattern = ".*mismatch0.*qs$",full.names = T)
# adapter.files.T44.mismatch1 = list.files("../data/FAX23635_pass_12b693ce_c74c2dc4/",pattern = ".*mismatch1.*qs$",full.names = T)
adapter.files.T44.mismatch2 = list.files("../data/FAX23635_pass_12b693ce_c74c2dc4/",pattern = ".*mismatch2.*qs$",full.names = T)
adapter.files.T44.mismatch3 = list.files("../data/FAX23635_pass_12b693ce_c74c2dc4/",pattern = ".*mismatch3.*qs$",full.names = T)
```

```r
adapter.list.T44.mismatch2 = mclapply(adapter.files.T44.mismatch2,function(file){
    df = as.data.frame(qread(file))
    type2 = str_replace(str_replace(str_replace(file,"../data/FAX23635_pass_12b693ce_c74c2dc4//FAX23635_pass_12b693ce_c74c2dc4_",""),".qs",""),".mismatch2","")
    df$type = type2
    df = df %>% select(-group_name) %>% mutate(fqname = fqnames_T44[group]) %>% select(-group)
    return(df)
},mc.cores = 20)
```

```r
adapter.list.T44.mismatch3 = mclapply(adapter.files.T44.mismatch3,function(file){
    df = as.data.frame(qread(file))
    type2 = str_replace(str_replace(str_replace(file,"../data/FAX23635_pass_12b693ce_c74c2dc4//FAX23635_pass_12b693ce_c74c2dc4_",""),".qs",""),".mismatch3","")
    df$type = type2
    df = df %>% select(-group_name) %>% mutate(fqname = fqnames_T44[group]) %>% select(-group)
    return(df)
},mc.cores = 20)
```
