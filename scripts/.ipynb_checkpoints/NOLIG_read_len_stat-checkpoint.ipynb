```r
fq = readDNAStringSet("..//T44_nolig_fastq/T44_nolig_all.fastq",format = "fastq")
```

```r
lencut = width(fq) %>% cut(.,breaks = c(seq(0,3000,100),Inf))
lencut.table = table(lencut)
lencut.df = data.frame(range = names(lencut.table),count = as.vector(lencut.table))
```

```r
adapter.files.nanopore.mismatch2 = list.files("../data/",pattern = ".*mismatch2.*qs$",full.names = T)
```

```r
nanopore_fqnames = read.table("../T44_nolig_fastq/T44_nolig_all.fastq.fqnames")
nanopore_fqnames = nanopore_fqnames$V1
```

```r
adapter.list.nanopore.mismatch2 = lapply(adapter.files.nanopore.mismatch2,function(file){
    df = as.data.frame(qread(file))
    type2 = str_replace(str_replace(str_replace(file,"../data//T44_nolig_all_",""),".qs",""),".mismatch2","")
    df$type = type2
    df = df %>% select(-group_name) %>% mutate(fqname = nanopore_fqnames[group]) %>% select(-group)
    return(df)
})
```

```r
components.nanopore = list("CIDcapture" = unique(adapter.list.nanopore.mismatch2[[1]]$fqname),"CIDcapture.rc" = unique(adapter.list.nanopore.mismatch2[[2]]$fqname),
"CIDoligo" = unique(adapter.list.nanopore.mismatch2[[3]]$fqname),"CIDoligo.rc" = unique(adapter.list.nanopore.mismatch2[[4]]$fqname),
"TSO" = unique(adapter.list.nanopore.mismatch2[[5]]$fqname),"TSO.rc" = unique(adapter.list.nanopore.mismatch2[[6]]$fqname))
```

```r
upsetlist.mx.nanopore = list_to_matrix(components.nanopore)
m1.nanopore = make_comb_mat(upsetlist.mx.nanopore)
m2.nanopore = m1.nanopore[comb_size(m1.nanopore)>=max(comb_size(m1.nanopore))*0.05]
upset.p = UpSet(m2.nanopore,comb_order = order(-comb_size(m2.nanopore)))
options(repr.plot.height = 4,repr.plot.width = 7)
upset.p
```
